```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
% matplotlib inline
# Always make it pretty.
plt.style.use('ggplot')
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df.head(10)
```

```python
response_rate = df['host_response_rate']
```

```python
rates = list(response_rate)
print(type(rates[0]))
```

```python
new_rate = []
for rate in rates:
    if type(rate) == str:
        rate_list = list(rate)[:-1]
        rate_perc = 0
        for i in range(len(rate_list)):
            rate_perc += int(rate_list[i])*(10**(len(rate_list)-i-1))
        new_rate.append(rate_perc)
    if type(rate) == float:
        new_rate.append(rate)
```

```python
df['host_response_rate'] = pd.DataFrame(new_rate)
```

```python
df.head(10)
```

```python
df.to_csv('cleaned_data.csv',index=False,sep=',')
```
